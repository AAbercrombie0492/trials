```python
import sys
sys.path.append('..')

from collections import Counter, OrderedDict

import seaborn as sbn
from matplotlib import pyplot as plt

from trials import Trials
```

Theoretical performance

```python
params = OrderedDict([('Control', 0.70), ('A', 0.60), ('B', 0.69), ('C', 0.72), ('D', 0.75)])
sources = OrderedDict([(label, sp.stats.bernoulli(p)) for label, p in params.items()])
```

```python
trial = Trials(params.keys())
samples = 10
total = 10000
metrics = {metric: [] for metric in ['lift', 'domination', 'frequentist domination', 'empirical lift']}

for k in range(0, total, samples):
    # Update
    data = OrderedDict([(label, source.rvs(samples)) for label, source in sources.items()])
    observations = OrderedDict([(label, (Counter(datum)[1], Counter(datum)[0])) for label, datum in data.items()])
    trial.update(observations)

    # Calculate metrics
    for key in metrics:
        metrics[key].append(trial.evaluate(key))
```

```python
for i, v in enumerate(sources):
    if v == 'Control':
        continue
        
    lifts = [lift.values[v] for lift in metrics['lift']]
    elifts = [p.values[v] for p in metrics['empirical lift']]
    ps = [p.values[v] for p in metrics['domination']]
    fps = [p.values[v] for p in metrics['frequentist domination']]

    fig, (p_plot, lift_plot) = plt.subplots(2, sharex=True)
    
    dpi = 118
    fig.set_size_inches(1600/dpi, 800/dpi, dpi=dpi)
    
    p_plot.set_xlim(0, total/samples)
    p_plot.set_ylim(0, 1)
    p_plot.plot(ps, label='Bayesian (Evan Miller) probability')
    p_plot.plot(fps, label='Frequentist (z-test) p-value')
    p_plot.hlines(0.95, 0, total/samples, color='grey', linestyle='--', label='0.95')
    p_plot.hlines(0.05, 0, total/samples, color='grey', linestyle='--', label='0.05')
    p_plot.set_xscale('symlog')

    lift_plot.plot(lifts, label='Bayesian expected lift')
    lift_plot.plot(elifts, label='Empirical lift')
    lift = (params[v]-params['Control'])/params['Control']
    lift_plot.hlines(lift, 0, total/samples, color='cyan', linestyle='--', label='True lift')
    lift_plot.set_xscale('symlog')
    
    if i == 1:
        p_plot.set_title('Domination')
        lift_plot.set_title('Lift')
        p_plot.legend(loc=5)
        lift_plot.legend(loc=1)
    
    fig.suptitle('Variation {}. Lift {:.2%}'.format(v, lift))
    fig.show()
    
```
